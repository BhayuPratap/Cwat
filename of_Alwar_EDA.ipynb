```python
import os
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
base_dir = "Alwar_precip/"

month1 = []
month2 = []
month3 = []
month4 = []

final_list = []

for i in os.listdir(base_dir):
    # print(base_dir + i)
    file = open(base_dir + i, "r")
    l = file.readlines()
    for i in range(1, len(l)):
        l[i] = l[i].strip()
        l[i] = l[i].split(" ")
        for j in range(len(l[i])):
            l[i][j] = float(l[i][j])
        # print(l[i])
        final_list.append(l[i])
    file.close()

print(final_list[0])
```

```python
final_array = np.array(final_list)
divided_array = np.array_split(final_array, 20)
arr = sum(divided_array)
print(arr.shape)
print(arr)
lst = arr.T.tolist()
print(len(lst))
print(lst[0])
print(len(lst[0]))
```
